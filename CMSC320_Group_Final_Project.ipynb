# **Age**

```python
# Age

columns_of_interest = [
    'Unnamed: 0',
    'Life satisfaction rating between 0 and 5 - Total',
    'Life satisfaction rating of 6 or 7 - Total',
    'Life satisfaction rating of 8, 9 or 10 - Total'
]

# Filtering the dataframe for age groups and satisfaction ratings
life_df_age_groups = life_df[columns_of_interest]
life_df_age_groups.columns = ['Age Group', 'Rating 0-5', 'Rating 6-7', 'Rating 8-10']

# Dropping rows that are not related to age groups
age_related_rows = ['Total, 15 years and over', '15 to 24 years', '25 to 54 years', '55 to 64 years', '65 years and over']
life_df_age_groups = life_df_age_groups[life_df_age_groups['Age Group'].isin(age_related_rows)]
life_df_age_groups.set_index('Age Group', inplace=True)

# Creating bar chart
life_df_age_groups.plot(kind='bar', title="Life Satisfaction Ratings by Age Group", color=['cyan', 'gold', 'pink'])
plt.ylabel("Life Satisfaction Rating (%)")
plt.xticks(rotation=45)
plt.show()
life_df_age_groups.describe()
```

```python
total_group = life_df_age_groups.loc["Total, 15 years and over"]

# Sample size for groups
n_total = n_age_group = 1

# Function to perform two-tailed Z-test on each category
def perform_z_test(group1, group2, n1, n2):
    z_scores = {}
    p_values = {}
    for rating in group1.index:
        mean_diff = group1[rating] - group2[rating]
        std_diff = np.sqrt((group1[rating]**2 / n1) + (group2[rating]**2 / n2))
        z_score = mean_diff / std_diff
        p_value = stats.norm.sf(abs(z_score)) * 2
        z_scores[rating] = z_score
        p_values[rating] = p_value
    return z_scores, p_values

# Loop through each age group and perform Z-test for each age group to the total average
results = []
for age_group in life_df_age_groups.index:
    if age_group != 'Total, 15 years and over':
        age_group_values = life_df_age_groups.loc[age_group]
        z_scores, p_values = perform_z_test(age_group_values, total_group, n_age_group, n_total)
        results.append((age_group, z_scores, p_values))

for result in results:
    age_group, z_scores, p_values = result
    print(f"\n{age_group}:")
    for rating in z_scores:
        print(f"{rating}: Z-score = {z_scores[rating]:.5f}, p-value = {p_values[rating]:.5f}")
```

**Conclusion**

H0: There is no signifcant difference between the ratings of the age group "15 to 24 years" (and every other age group range from the data) to the total age group, "Total, 15 years and over"

HA: There is a signifcant difference between the ratings of the age group "15 to 24 years" (and every other age group range from the data) to the total age group, "Total, 15 years and over"

After performing a Z-test on every age group subset compared to the total life satisfaction of all ages, it was found that there is no signficant difference between any ratings to the population mean ratings. Since none of the p-values for each age subset and rating subset was less than .05, this means that there were no signifcant difference between the ratings of each age group subset to the overall ratings of all ages. Thus, the null hypothesis is true in all cases.

# **Location**

```python
# Location

from scipy.stats import chi2_contingency

# Filter columns for location analysis
columns_of_interest_location = [
    'Unnamed: 0',
    'Life satisfaction rating between 0 and 5 - Total',
    'Life satisfaction rating of 6 or 7 - Total',
    'Life satisfaction rating of 8, 9 or 10 - Total'
]

# Filter columns for rural vs urban areas and satisfaction ratings
life_df_location = life_df[columns_of_interest_location]
life_df_location.columns = ['Location', 'Rating 0-5', 'Rating 6-7', 'Rating 8-10']

# Filter rows for just urban and rural
filtered_df = life_df_location[life_df_location['Location'].isin(['Urban areas 20', 'Rural areas 20'])]

contingency_table = pd.DataFrame({
    'Urban': filtered_df[filtered_df['Location'] == 'Urban areas 20'].iloc[0, 1:],
    'Rural': filtered_df[filtered_df['Location'] == 'Rural areas 20'].iloc[0, 1:]
})

# Perform the Chi-Squared test
chi2, p, dof, expected = chi2_contingency(contingency_table.T)

# bar chart for data visualization
contingency_table.plot(kind='bar', title="Life Satisfaction Ratings by Location (Chi-Squared Test)", color=['blue', 'orange'])
plt.ylabel("Life Satisfaction Rating (%)")
plt.xticks(rotation=0)
plt.show()

# Output the Chi-Squared statistic and p-value
print(f"Chi-Squared Statistic: {chi2}, p-value: {p}")
```

**Conclusion**

Null Hypothesis (H0): There is no significant difference in the distribution of life satisfaction ratings between urban and rural areas.

Alternative Hypothesis (HA): There is a significant difference in the distribution of life satisfaction ratings between urban and rural areas.

Based on the Chi-Squared test results, with a Chi-Squared statistic of 2.22 and a p-value of 0.329, we can conclude that there is no statistically significant difference in life satisfaction ratings between urban and rural areas. The p-value, being higher than the common threshold of 0.05, indicates that we fail to reject the null hypothesis, meaning that the distribution of life satisfaction ratings across the categories (0-5, 6-7, 8-10) is similar for both urban and rural locations. Any differences observed in the data are likely due to random variation rather than a true underlying difference. Therefore, life satisfaction appears to be consistent across these environments, with no strong evidence to suggest that location significantly impacts overall satisfaction levels.

# **Education**

The following section is on hypothesis testing regarding education level and life satisfaction. The first bar graph simply displays life satisfaction ratings based on the three education levels: No certificate, diploma or degree, Secondary (high) school diploma or equivalency certificate, and Postsecondary certificate, diploma or degree. This graph is just for visualizing the data. The next segment shows an ANOVA test hypothesis analysis on the data in search of correlations between education level and life satisfaction.

```python
# Education

# Extractin the relevant columns for analysing the relationship between education
# level and life satisfaction.
education_levels = ['No certificate, diploma or degree',
                   'Secondary (high) school diploma or equivalency certificate',
                   'Postsecondary certificate, diploma or degree']
satisfaction_ratings = ['Life satisfaction rating between 0 and 5 - Total',
                       'Life satisfaction rating of 6 or 7 - Total',
                       'Life satisfaction rating of 8, 9 or 10 - Total']

# Filter life_df for the selected education levels.
filtered_df = life_df[life_df['Unnamed: 0'].isin(education_levels)]

# Colors for the graph.
colors = ['turquoise', 'lemonchiffon', 'mediumvioletred']

# Create a bar plot for basic visualization of the data:
filtered_df.set_index('Unnamed: 0')[satisfaction_ratings].plot(kind='bar', color=colors)
plt.title('Life Satisfaction Ratings by Educational Level')
plt.xlabel('Educational Level')
plt.ylabel('Life Satisfaction (%)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Key', bbox_to_anchor=(1, 1))
plt.show()
```

```python
# The oneway function requires an array of input.
# The following for loop creates an array for each category's data.
data_for_anova = [filtered_df[filtered_df['Unnamed: 0'] == level][satisfaction_ratings].values.flatten()
                  for level in education_levels]

# ANOVA testing:
_, pvalue = stats.f_oneway(*data_for_anova)

print(f"p-value: {pvalue}")

# Creating a box and whisker plot to display the means of each group.
plt.xticks(rotation=45, ha='right')
sns.boxplot(x='Unnamed: 0', y='value', data=pd.melt(filtered_df[satisfaction_ratings + ['Unnamed: 0']], id_vars=['Unnamed: 0']), order=education_levels, color='mediumpurple')
plt.xlabel('Education Level')
plt.ylabel('Life Satisfaction Rating')
plt.title('Life Satisfaction Ratings by Education Level (ANOVA Test)')
plt.show()
```

**Conclusion**

Originally, we hypothesized that there would be a positive correlation between higher education levels and increased life satisfaction. However, after running the ANOVA test on the data, we got a p value of 0.999. This is significantly greater than our alpha value/significance level of 0.05, meaning there is no significant difference between the mean life satisfaction ratings of the 3 eduction levels: No certificate, diploma or degree, Secondary (high) school diploma or equivalency certificate, and Postsecondary certificate, diploma or degree. Although this is a preliminary test, it indicates that education level may not have as significant of an impact on life satisfaction as we originally thought, disproving our hypothesis.
